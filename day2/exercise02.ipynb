```python
def is_even(a):
    """
    this function returns True if the number is even otherwise False
    """
    result = a % 2
    if result == 1:
        return False
    else:
        return True
        
is_even(10)        
```

```python
def is_even_2(a):
    if a % 2 == 1:
        return False
    else:
        return True

is_even_2(11)
```

```python
def is_even_4(a):
    return (a % 2 == 0) # returns True if the result is 0 otherwise returns False
```
