# for-loops

```python
# ex 1
num = input("insert a number")
num = float(num) # the input function returns a string
if num > 0:
    print("positive")
elif num == 0:
    print("zero")
else:
    print("negative")
```

```python
# ex 2

age = input("insert your age")
age = int(age)
if age < 13:
    print("Child")
elif age >= 13 and age <= 19:
    print("Teenager")
elif age >= 20 and age <= 65:
    print("Adult")
else:
    print("Senior")
```

```python
# ex 3
grade = float(input("insert your grade vote"))
if grade >= 90 and grade <= 100:
    print("A")
elif grade >= 80 and grade <= 89:
    print("B")
elif grade >= 70 and grade <= 79:
    print("C")
elif grade >= 60 and grade <= 69:
    print("D")
else:
    print("E")
```

```python
# ex 4
v1 = input("boolean variable 1:")
if v1.lower() == "true":
    v1 = True
else: 
    v1 = False
v2 = input("boolean variable 2:")
if v2.lower() == "true":
    v2 = True
else: 
    v2 = False

print("v1 and v2: ",v1 and v2)
print("v1 or v2: ",v1 or v2)
print("not v1: ",not v1)
```
