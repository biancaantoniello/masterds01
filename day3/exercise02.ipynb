# TKINTER

```python
import tkinter as tk
```

```python
root = tk.Tk()
```

```python
root.title("My first Tkinter Window") # title for the window
```

```python
welcome_label = tk.Label(root, text = "Welkome to Tkinter") # 1st widget
```

```python
welcome_label.pack()
```

```python
def on_button_click():
    welcome_label.config(text = "Button clicked") # modifies the text element
```

```python
click_me_button = tk.Button(root, text = "click me!", command = on_button_click) # 2nd widget
```

```python
click_me_button.pack()
```

```python
root.mainloop() # create the window (to run at the end)
```
