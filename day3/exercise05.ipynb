# MATPLOTLIB

```python
#!pip install matplotlib
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,3,5,1,11]

plt.plot(x,y)
plt.title("sample plot")
plt.xlabel("days")
plt.ylabel("rain mm")
plt.show()
```

# BAR CHART

```python
categories = ["Company A","Company B", "Company C", "Company D"]

values = [20,50,10,40]

colors = ["green", "blue", "red", "pink"]

plt.bar(categories, values, color = colors)
plt.title("bar chart")
plt.xlabel("Northwest companies")
plt.ylabel("revenues")
plt.xticks(rotation = 45)
plt.show()
```
